```python
import sys
import os

# Adds the parent directory to the python path
sys.path.append(os.path.abspath('../'))

from src.physics.dynamics import HSRSolver
from src.physics.simulation import MonteCarloRunner
from src.visualization.plots import plot_hsr_results

# 1. Initialize Engine
solver = HSRSolver()
runner = MonteCarloRunner(solver)

# 2. Run the Stochastic Search (Kinney/Efstathiou Method)
# We run a large batch because most random models fail to end inflation
# n_min=40, n_max=70 defines the 'Graceful Exit' window
print("Exploring the Flow Landscape...")
results = runner.run_batch(n_sims=50000, m_order=10, method='stochastic')#increase n_sims for more points

# 3. Visualization
# This plot shows the wide 'chaotic' scatter typical of Efstathiou's landscape
plot_hsr_results(results[:, 0], results[:, 1], m_order=10)
```
